```python
import java.util.Random;
import java.util.Scanner;

public class MonsterHuntGame {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        Random random = new Random();

        int playerHP = 100;
        int monsterHP = 50;
        int playerAttack = 15;
        int monsterAttack = 10;

        System.out.println("⚔️ 몬스터 사냥 게임 시작! ⚔️");

        while (playerHP > 0 && monsterHP > 0) {
            System.out.println("\n[플레이어 HP: " + playerHP + "] [몬스터 HP: " + monsterHP + "]");
            System.out.println("1. 공격하기  2. 도망가기");
            System.out.print("선택: ");
            int choice = scanner.nextInt();

            if (choice == 1) {
                int playerDamage = random.nextInt(playerAttack) + 1;
                int monsterDamage = random.nextInt(monsterAttack) + 1;

                monsterHP -= playerDamage;
                System.out.println("\n💥 몬스터에게 " + playerDamage + "의 피해를 입혔다!");

                if (monsterHP > 0) {
                    playerHP -= monsterDamage;
                    System.out.println("😈 몬스터가 반격하여 " + monsterDamage + "의 피해를 입었다!");
                }
            } else if (choice == 2) {
                System.out.println("🏃‍♂️ 도망쳤다! 게임 종료.");
                break;
            } else {
                System.out.println("⚠️ 올바른 선택을 입력하세요!");
            }
        }

        if (playerHP <= 0) {
            System.out.println("😭 당신은 패배했습니다...");
        } else if (monsterHP <= 0) {
            System.out.println("🎉 몬스터를 처치했다! 승리!");
        }

        scanner.close();
    }
}
```

```python
import java.util.Random;
import java.util.Scanner;

class Monster {
    String name;
    int hp;
    int attack;

    public Monster(String name, int hp, int attack) {
        this.name = name;
        this.hp = hp;
        this.attack = attack;
    }

    public boolean isAlive() {
        return hp > 0;
    }
}

public class MonsterBattleGame {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        Random random = new Random();

        Monster monster = new Monster("드래곤", 100, 20);
        int playerHP = 120;
        int playerAttack = 25;

        System.out.println("🔥 몬스터 배틀 시작! 🔥");

        while (playerHP > 0 && monster.isAlive()) {
            System.out.println("\n[플레이어 HP: " + playerHP + "] [" + monster.name + " HP: " + monster.hp + "]");
            System.out.println("1. 공격하기  2. 방어하기  3. 도망가기");
            System.out.print("선택: ");
            int choice = scanner.nextInt();

            if (choice == 1) {
                int playerDamage = random.nextInt(playerAttack) + 1;
                monster.hp -= playerDamage;
                System.out.println("💥 " + monster.name + "에게 " + playerDamage + "의 피해를 입혔다!");

                if (monster.isAlive()) {
                    int monsterDamage = random.nextInt(monster.attack) + 1;
                    playerHP -= monsterDamage;
                    System.out.println("🔥 " + monster.name + "의 반격! " + monsterDamage + "의 피해를 입었다!");
                }
            } else if (choice == 2) {
                int reducedDamage = random.nextInt(10) + 1;
                playerHP -= reducedDamage;
                System.out.println("🛡 방어 성공! 피해를 " + reducedDamage + "만큼 줄였다!");
            } else if (choice == 3) {
                System.out.println("🏃‍♂️ 도망쳤다! 게임 종료.");
                break;
            } else {
                System.out.println("⚠️ 올바른 선택을 입력하세요!");
            }
        }

        if (playerHP <= 0) {
            System.out.println("😵 당신은 패배했습니다...");
        } else if (!monster.isAlive()) {
            System.out.println("🎉 " + monster.name + "를 처치했다! 승리!");
        }

        scanner.close();
    }
}
```
